```python
cor = np.random.choice(np.linspace(-1, 1, 100), 100).reshape(10,10)
```

# Project 2 

```python
# 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，
# 並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
```

```python
# 生成資料
data = np.random.choice(np.linspace(-1, 1), 3000).reshape(1000, 3)
```

```python
# 將資料製作成girdplot
tgrid = sns.PairGrid(pd.DataFrame(data))
# 繪圖
tgrid.map_upper(plt.scatter)
tgrid.map_diag(plt.hist)
tgrid.map_lower(sns.kdeplot)
```
