```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
app = pd.read_csv('C:/Users/hsu/Desktop/application_train.csv')
```

```python
app.shape
```

```python
# 計算任意欄位的平均數及標準差
print('平均值 = ', app['AMT_GOODS_PRICE'].mean())
print('標準差 = ', app['AMT_GOODS_PRICE'].std())
print('---四分位數---')
print(app['AMT_GOODS_PRICE'].quantile(np.arange(0,1.25,0.25)))
```

```python
## 畫出任意欄位的直方圖
# 選取資料欄位類型為數值型資料
# 
# ps：不選取(int)資料，主要在於此種資料表面雖為數值型資料，但可能其隱含意義可能為「類別型資料(偏以順序類型居多)」，
# 為顧及分析順利，因選「float64」資料為主。

floatlis = []
for i in app.columns:
    if app[i].dtype in ['float64']:
        floatlis.append(i)
```

```python
# 繪製資料中所有「float64」之直方圖
plt.figure(figsize=(20,20))
for i in range(len(floatlis)):
    plt.subplot(9, 8, i + 1)
    plt.hist(x = app[floatlis[i]])
    plt.title(floatlis[i], size = 8)
```
