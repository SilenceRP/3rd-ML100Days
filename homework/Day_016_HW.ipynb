```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 載入資料
app = pd.read_csv('C:/Users/hsu/Desktop/application_train.csv')
```

```python
# 取出繪圖欄位
drawapp = app[['DAYS_BIRTH','TARGET']]
```

```python
# 劃定範圍
rge = np.linspace(20,70,11)
print(rge)
```

```python
# 另設欄位
drawapp['agerge'] = pd.cut(drawapp['age'], rge, labels=['20-25','25-30','30-35','35-40','40-45',
                                                        '45-50','50-55','55-60','60-65','65-70'])
```

```python
# 資料檢視
drawapp['agerge'].value_counts()
```

```python
# 繪圖1
plt.figure(figsize=(10, 10))
for i in drawapp['agerge'].unique():
    sns.kdeplot(drawapp.loc[drawapp['agerge'] == i, 'age'])
```

```python
# 繪圖2
plt.figure(figsize=(15,10))
for i in range(len(drawapp['agerge'].unique())):
    plt.subplot(2,5,i+1)
    sns.kdeplot(drawapp.loc[drawapp['agerge'] == drawapp['agerge'].unique()[i],'age'])
    plt.title(drawapp['agerge'].unique()[i])
```

```python
# 以年齡區間為 x, target 為 y 繪製 barplot (直方圖)
plt.hist(drawapp.loc[drawapp['TARGET'] == 0, 'age'])
plt.hist(drawapp.loc[drawapp['TARGET'] == 1, 'age'])
```

```python
# 以年齡區間為 x, target 為 y 繪製 barplot (長條圖)
plt.figure(figsize=(10,10))
plt.bar(drawapp.loc[drawapp['TARGET'] == 0, 'agerge'].unique(), 
        drawapp.loc[drawapp['TARGET'] == 0,'agerge'].value_counts(),
        align='edge',
       width=0.35)
plt.bar(drawapp.loc[drawapp['TARGET'] == 1, 'agerge'].unique(), 
        drawapp.loc[drawapp['TARGET'] == 1,'agerge'].value_counts(),
        align='edge',
       width=-0.35)
```
