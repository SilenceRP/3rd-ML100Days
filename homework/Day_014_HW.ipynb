```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 弱相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

# 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x, y)
```

```python
# 將分布畫出來看看吧
plt.scatter(x, y)
```

### 正相關

```python
# 隨機生成 1000 個介於 0~50 的數 x
x = np.random.randint(0, 50, 1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
y = x + np.random.normal(0, 10, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x, y)
```

```python
# 再看看正相關的 x,y 分布
plt.scatter(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 模擬正相關數值
px = np.random.randint(0, 50, 1000)
py = px + np.random.normal(0, 10, 1000)

# 模擬負相關數值
nx = np.random.randint(0, 50, 1000)
ny = nx + np.random.normal(0, 10, 1000)
ny1 = ny * -1
```

```python
# 正、負相關係數
print('正相關係數 ')
print(np.corrcoef(px, py))
print('\n')
print('負相關係數 ')
print(np.corrcoef(nx, ny1))
```

```python
# 繪圖
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(px, py)
plt.title('Positive Corr')

plt.subplot(1,2,2)
plt.scatter(nx, ny1)
plt.title('Negative Corr')
```
